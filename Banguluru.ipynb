```python
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder_area_type = LabelEncoder()
```

```python
data.area_type = encoder_area_type.fit_transform(data.area_type)
```

```python
data.head()
```

```python
encoder_availability = LabelEncoder()
data.availability = encoder_availability.fit_transform(data.availability)
```

```python
encoder_location = LabelEncoder()
data.location = encoder_location.fit_transform(data.location)
```

```python
data.head()
```

```python
encoder_size = LabelEncoder()
data.size = encoder_size.fit_transform(data.size)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='most_frequent')
```

```python
data.society = imputer.fit_transform(data[['society']])
```
